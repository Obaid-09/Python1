GROUP BY AND AGGREGATING

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Sufyaan ANS7\OneDrive\Documents\UG Branchwise Placement Stats 2024-25.xlsx", sheet_name = 'EEE')
df
```

```python
items = ['Name','Roll No.','Gender','CGPA','Company','Type of offer','Job profile','CTC(LPA)']
df = df[items]
df
```

```python
new_items = ['CGPA','Company','CTC(LPA)']
df2 = df[new_items]
df2
```

```python
group_by = df2.groupby('Company')
```

```python
group_by.mean()
```

```python
df2.groupby('Company').count()
```

```python
df2.groupby('Company').min()
```

```python
df2.groupby('Company').max('CGPA')
```

```python
df2.groupby('Company').max('CTC(LPA)')
```

```python
df2.groupby('Company').sum()
```

```python
df2.groupby('Company').agg({'CTC(LPA)' : ['mean', 'max','min', 'count', 'sum'], 'CGPA' : ['mean', 'max','min', 'count', 'sum']})
```

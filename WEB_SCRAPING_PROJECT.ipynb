```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = "wikitable plainrowheaders sortable")
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
titles = soup.find_all('tr')[7]
```

```python
print(titles)
```

```python
table_titles = [title.text.strip() for title in titles]

cleaned_table_titles = [title for title in table_titles if title.strip() != '']
print(cleaned_table_titles)
```

```python
cleaned_table_titles[7] = 'Attendance'
```

```python
cleaned_titles = [title for title in cleaned_table_titles if title.strip() != 'Season']
```

```python
print(cleaned_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = cleaned_titles)

df
```
